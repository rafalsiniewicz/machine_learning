# K-means clustering - example I

```python
from sklearn.datasets.samples_generator import make_blobs 
 
help(make_blobs)
```

#### Ex.1

#### Parameters:
- n_samples : int or array-like, optional (default=100)
- n_features : int, optional (default=2)
- centers : int or array of shape [n_centers, n_features], optional (default=None)
- cluster_std : float or sequence of floats, optional (default=1.0)
- center_box : pair of floats (min, max), optional (default=(-10.0, 10.0))
- shuffle : boolean, optional (default=True)
- random_state : int, RandomState instance or None (default)

#### Generated output:
- X : array of shape [n_samples, n_features]
- y : array of shape [n_samples]

#### Ex.2

```python
X, y = make_blobs(n_samples=120, centers=4, n_features=2, random_state=0, cluster_std=0.4)
```

#### Ex.3

```python
print("X.shape = ", X.shape)
print("y.shape = ", y.shape)
```

#### Ex.4

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.show()
```

#### Ex.5

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0, init='random')
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

#### Ex.6 

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```

#### Ex.7

```python
from base import dunn
import numpy as np
k_list = [np.append(X[i], X[i]) for i,x in enumerate(y_pred) if x==0] 
# k_list

print("inertia = ", kmeans.inertia_)
```

```python
ks = [np.empty((0,2)) for i in range(4)]
k_list = [k for k in ks]

for i in range(y_pred.shape[0]):
    kl = y_pred[i]
    arr = k_list[kl]
    arr = np.append(arr,X[i,:].reshape(1,2),0)
    k_list[kl] = arr

print("Dunn index = ", dunn(k_list))
```

# Elbow method

```python
k_list = [i for i in range(1,20)]
inertia_list = []
```

```python
for i in k_list:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_)
```

```python
plt.scatter(k_list, inertia_list)
plt.xticks(k_list)
plt.show()
```
