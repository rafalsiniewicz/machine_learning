```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
# load the famous iris data
irisRaw = load_iris()
print(irisRaw)
 
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'] + ['class'])

iris.loc[iris['target'] == 0, 'class'] = 'setosa'
iris.loc[iris['target'] == 1, 'class'] = 'versicolor'
iris.loc[iris['target'] == 2, 'class'] = 'virginica'
```

```python
#1
print(iris['sepal length (cm)'][0])
print(iris['sepal width (cm)'][0])
iris.head()
```

```python
#2 
print("number of rows = ", len(iris))
print("number of columns = ", len(iris.columns))
print(iris.info())
```

```python
#3
print(iris['sepal length (cm)'].describe())
print(iris['sepal width (cm)'].describe())
print(iris['petal length (cm)'].describe())
print(iris['petal width (cm)'].describe())
print(iris['target'].describe())

grouped = iris.groupby('target')
grouped.first()
```

```python
#4
iris.head()
```

```python
#5
print(len(iris))
dropped = iris.dropna()
print(len(dropped))
if len(dropped) == len(iris):
    print("no nan values")
else:
    print("{} nan values".format(len(iris) - len(dropped)))
```

```python
#6
print(iris.sort_index(axis=1))
#second column
print(iris.sort_values('sepal width (cm)'))
```

```python
#7
#third column
print(iris.sort_values('petal length (cm)'))
#maximum petal length
print(iris['petal length (cm)'].max())
#minimum petal length
print(iris['petal length (cm)'].min())
#index of maximum petal length
print(iris['petal length (cm)'].idxmax())
#index of minimmum petal length
print(iris['petal length (cm)'].idxmin())
```

```python
#8
print(iris.std(axis = 0, skipna = True))
```

```python
#9
av_len = iris.mean(axis = 0, skipna = True)[0]
#rows for which 'sepal length (cm)' is greater than mean
print(iris[iris['sepal length (cm)'] > av_len])
```

```python
#10
import matplotlib.pyplot as plt

class_names = iris['class'].unique()


for j in range(len(class_names)):
    fig,a =  plt.subplots(4,1, figsize=(15,15))
    fig.suptitle(class_names[j])
    for i in range(len(iris.columns) - 2):
        data = iris[iris['class'] == class_names[j]][iris.columns[i]]
        a[i].hist(data, bins=10)
        plt.setp(a[i], xlabel=iris.columns[i])
        plt.setp(a[i], ylabel='amount')
```
